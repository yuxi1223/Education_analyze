```python
#資料匯入及處理

#資料匯入
import pandas as pd
url = "https://ws.moe.edu.tw/001/Upload/4/relfile/0/4758/b3052b5f-c0b0-4484-beda-c43a56fd317a.csv"
df = pd.read_csv(url,thousands=',')

df
```

```python
#顯示每個年度退學人數
df1 = df[["學年度","學校名稱","學制班別","性別","在學學生數","退學人數小計"]]
df_105=df1[df1["學年度"]==105]
df_106=df1[df1["學年度"]==106]
df_107=df1[df1["學年度"]==107]
```

```python
#生成html檔案-篩選欄位過的資料（"學年度","學校名稱","學制班別","性別","在學學生數","退學人數小計"）
webdata=df1.to_html()
print(webdata)
with open ("webdata.html","w") as file:
  file.write(webdata)

```

```python
#台灣105-107個年度總和退學原因-長條圖

import pandas as pd
import matplotlib.pyplot as plt

condiction = ["因學業成績退學人數", "因操行成績退學人數", "因志趣不合退學人數", "因逾期未註冊退學人數", "因休學逾期未復學退學人數", "因懷孕退學人數", "因育嬰退學人數", "因病退學人數", "因工作需求退學人數", "其他(不含死亡)退學人數"]

total_counts = df[condiction].sum()

colors = ['lightblue', 'yellow', 'lightgreen', 'orange','salmon', 'pink', 'purple', 'cyan']

# 設定圖表大小
plt.figure(figsize=(16,6))

plt.bar(total_counts.index, total_counts, color=colors)

plt.title("105~107學年度台灣各所大學退學原因", fontsize=16)
plt.xlabel("退學原因", fontsize=16)
plt.ylabel("退學人數", fontsize=16)
plt.xticks(rotation=90)

for i, val in enumerate(total_counts):
    plt.text(i, val, str(val), color='black', ha='center', va='bottom')

plt.show()
```

```python
#105年退學人數最高大學-中國文化大學（學制分類）
import pandas as pd
import matplotlib.pyplot as plt

condicition_boy = (df_105["學校名稱"]=="中國文化大學")& (df_105["性別"]=="男")
condicition_girl = (df_105["學校名稱"]=="中國文化大學")& (df_105["性別"]=="女")

x1=df_105[condicition_girl].groupby(["學制班別"])["退學人數小計"].sum()
x2=df_105[condicition_boy].groupby(["學制班別"])["退學人數小計"].sum()


# 設定顏色
color1 = 'lightskyblue'
color2 = 'salmon'

# 設定圖表大小
plt.figure(figsize=(16,6))

# 繪製堆疊圖並設定顏色和數據標籤
plt.bar(x1.index, x1, color=color2, label='女')
plt.bar(x2.index, x2, bottom=x1, color=color1, label='男')

# 在圖中顯示數據
for i, val in enumerate(x1):
    plt.text(i, val/2, str(val), color='navy', ha='center', va='center', fontweight='bold', fontsize=16)
for i, val in enumerate(x2):
    plt.text(i, val/2 + x1[i], str(val), color='navy', ha='center', va='center', fontweight='bold', fontsize=16)

plt.title("105年退學人數最高大學-中國文化大學退學人數堆疊圖", fontsize=16)
plt.xlabel("學制班別", fontsize=16)
plt.ylabel("退學人數", fontsize=16)

plt.legend()
plt.show()
```
